```python
import pandas as pd

# Load the air quality dataset
dataset_path = "cpcb_dly_aq_tamil_nadu-2014.csv"  # Replace with the actual path to your dataset
data = pd.read_csv(dataset_path)
```

```python
# Inspect the data
print(data.head())
print(data.info())
print(data.describe())
```

```python
data.dropna(inplace=True)  # Remove rows with missing values
```

```python
data['Sampling Date'] = pd.to_datetime(data['Sampling Date'])  # Convert date column to datetime
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['Year'] = data['Sampling Date'].dt.year
data['Month'] = data['Sampling Date'].dt.month
```
